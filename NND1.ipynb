```python
import numpy as np

import tensorflow as tf



x = np.array([[1, 0, 1]])



w_input_hidden = np.array([

  [0.5, 0.4],

  [-0.2, 0.1],

  [0.3, -0.5]

])



w_hidden_output = np.array([0.7, 0.2])
hidden_layer = np.matmul(x, w_input_hidden)

raw_output = np.matmul(hidden_layer, w_hidden_output)

final_output = tf.sigmoid(raw_output)

print(hidden_layer)

print(final_output)
```

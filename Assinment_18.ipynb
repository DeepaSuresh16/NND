```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Load image using OpenCV
image = cv2.imread('Bird.jpg')   # Replace with your image file name

# Check if image loaded successfully
if image is None:
    print("Error: Image not found.")
    exit()

# -------- Display using OpenCV --------
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# -------- Display using Matplotlib --------
# Convert BGR to RGB (because OpenCV loads image in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Image using Matplotlib")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load color image
image = cv2.imread('Bird.jpg')   # Replace with your image name

if image is None:
    print("Error: Image not found.")
    exit()

# Convert BGR to RGB for correct display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display both images
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()

# ---------------------------
# Compare Pixel Intensity
# ---------------------------

# Choose one pixel location (example: center pixel)
row = image.shape[0] // 2
col = image.shape[1] // 2

# Color pixel (BGR)
color_pixel = image[row, col]

# Gray pixel
gray_pixel = gray_image[row, col]

print("Pixel Location:", (row, col))
print("Color Pixel Value (B, G, R):", color_pixel)
print("Grayscale Pixel Intensity:", gray_pixel)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('Bird.jpg')   # Replace with your image name

# Check if image loaded successfully
if image is None:
    print("Error: Image not found.")
    exit()

# Get original dimensions
h, w = image.shape[:2]

# Resize images
half_image = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
double_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Convert to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
half_rgb = cv2.cvtColor(half_image, cv2.COLOR_BGR2RGB)
double_rgb = cv2.cvtColor(double_image, cv2.COLOR_BGR2RGB)

# -------- Display Separately --------

# Original
plt.figure(figsize=(w/100, h/100))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Half Size
plt.figure(figsize=(w/200, h/200))
plt.imshow(half_rgb)
plt.title("Half Size Image")
plt.axis("off")
plt.show()

# Double Size
plt.figure(figsize=((w*2)/100, (h*2)/100))
plt.imshow(double_rgb)
plt.title("Double Size Image")
plt.axis("off")
plt.show()

# Print dimensions
print("Original Dimensions:", image.shape)
print("Half Dimensions:", half_image.shape)
print("Double Dimensions:", double_image.shape)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('Bird.jpg')   # Replace with your image name

if image is None:
    print("Error: Image not found.")
    exit()

# Rotate image
rotate_90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotate_180 = cv2.rotate(image, cv2.ROTATE_180)
rotate_270 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Convert BGR to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
r90_rgb = cv2.cvtColor(rotate_90, cv2.COLOR_BGR2RGB)
r180_rgb = cv2.cvtColor(rotate_180, cv2.COLOR_BGR2RGB)
r270_rgb = cv2.cvtColor(rotate_270, cv2.COLOR_BGR2RGB)

# Display images
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(r90_rgb)
plt.title("90 Degree Rotation")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(r180_rgb)
plt.title("180 Degree Rotation")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(r270_rgb)
plt.title("270 Degree Rotation")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload file
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Read image
image = cv2.imread(image_path)

if image is None:
    print("Error loading image")
else:
    # Flip horizontally (left  right)
    flip_horizontal = cv2.flip(image, 1)

    # Flip vertically (top  bottom)
    flip_vertical = cv2.flip(image, 0)

    # Flip both directions (optional)
    flip_both = cv2.flip(image, -1)

    # Convert BGR to RGB for display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    h_flip_rgb = cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB)
    v_flip_rgb = cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB)
    both_flip_rgb = cv2.cvtColor(flip_both, cv2.COLOR_BGR2RGB)

    # Display images
    plt.figure(figsize=(10,8))

    plt.subplot(2,2,1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(2,2,2)
    plt.imshow(h_flip_rgb)
    plt.title("Horizontal Flip")
    plt.axis("off")

    plt.subplot(2,2,3)
    plt.imshow(v_flip_rgb)
    plt.title("Vertical Flip")
    plt.axis("off")

    plt.subplot(2,2,4)
    plt.imshow(both_flip_rgb)
    plt.title("Horizontal + Vertical Flip")
    plt.axis("off")

    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('Bird.jpg')

if image is None:
    print("Error loading image")
    exit()

# Define crop coordinates
x1, y1 = 80, 40
x2, y2 = 250, 160

# Crop
cropped = image[y1:y2, x1:x2]

# Draw rectangle on original image
image_with_box = image.copy()
cv2.rectangle(image_with_box, (x1, y1), (x2, y2), (0,255,0), 2)

# Convert to RGB
orig_rgb = cv2.cvtColor(image_with_box, cv2.COLOR_BGR2RGB)
crop_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(orig_rgb)
plt.title("Original Image (Crop Area Shown)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(crop_rgb)
plt.title("Cropped Image")
plt.axis("off")

plt.show()

print("Original Shape:", image.shape)
print("Cropped Shape:", cropped.shape)
```

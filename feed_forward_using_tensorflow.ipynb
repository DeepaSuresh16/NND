```python
import tensorflow as tf
import numpy as np

x = tf.constant([0.5, 0.2, 0.1], dtype=tf.float32)

w = tf.constant([[0.1, 0.4],
                 [0.2, 0.3],
                 [0.3, 0.6]], dtype=tf.float32)
b = tf.constant([0.01, 0.02], dtype=tf.float32)

z = tf.matmul(tf.reshape(x, (1, 3)), w) + b
a = tf.sigmoid(z)

w2 = tf.Variable([0.2, 0.3], dtype=tf.float32)
b2 = tf.constant([0.05], dtype=tf.float32)

z2 = tf.reduce_sum(a * w2) + b2
a2 = tf.sigmoid(z2)

y = tf.constant(1.0)
loss = (y - a2) ** 2

lr = 0.1
with tf.GradientTape() as tape:
    z2 = tf.reduce_sum(a * w2) + b2
    a2 = tf.sigmoid(z2)
    loss = (y - a2) ** 2

grad_w2 = tape.gradient(loss, w2)
w2.assign_sub(lr * grad_w2)

print("Final Output:", a2.numpy())
print("Loss:", loss.numpy())
print("Updated w2:", w2.numpy())

```
